# Conditional Probability

Conditional probability measures the probability of an event occurring given that another event has already occurred.

The conditional probability of event A given event B is denoted as:
\[
P(A|B) = \frac{P(A \cap B)}{P(B)}
\]
where:
- \(P(A|B)\) is the probability of A given B.
- \(P(A \cap B)\) is the probability of both A and B happening.
- \(P(B)\) is the probability of event B.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Example: Rolling Two Dice

Suppose we roll two dice. Let:
- Event A be "the sum of the dice is 7."
- Event B be "the first die is 3."

What is the probability of Event A given Event B?


```python
# Total possible outcomes for two dice
total_outcomes = 6 * 6

# Event B: The first die is 3 (6 possible outcomes for the second die)
event_B_outcomes = 6

# Event A ∩ B: The first die is 3 and the sum is 7 (only one possibility: (3, 4))
event_A_and_B_outcomes = 1

# Conditional Probability: P(A|B) = P(A ∩ B) / P(B)
P_A_given_B = event_A_and_B_outcomes / event_B_outcomes

print(f"P(A|B) = {P_A_given_B:.2f}")
```

## Real-World Example: Iris Dataset

Let's calculate the probability of a flower being of species Setosa given that its petal length is less than 2 cm.


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Event A: Species is Setosa (species 0)
event_A = df['species'] == 0

# Event B: Petal length < 2 cm
event_B = df['petal length (cm)'] < 2

# Calculate P(A|B)
P_A_given_B = np.mean(event_A & event_B) / np.mean(event_B)
print(f"P(Species is Setosa | Petal length < 2 cm) = {P_A_given_B:.2f}")
```

The below code plots two histograms:

- One for the petal lengths of Setosa species (in blue).
- One for non-Setosa species (in green).

The vertical red line shows the cutoff at 2 cm for petal length, and the plot title includes the calculated conditional probability:

\[
P(\text{Setosa} | \text{Petal length} < 2 \, \text{cm})
\]


```python
# Importing necessary libraries for visualization
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Event A: Species is Setosa (species 0)
event_A = df['species'] == 0

# Event B: Petal length < 2 cm
event_B = df['petal length (cm)'] < 2

# Calculate P(A|B)
P_A_given_B = np.mean(event_A & event_B) / np.mean(event_B)

# Plot the distribution of petal length to visualize the conditional probability
plt.figure(figsize=(8, 6))
plt.hist(df['petal length (cm)'][event_A], bins=20, alpha=0.6, label="Setosa", color="blue")
plt.hist(df['petal length (cm)'][~event_A], bins=20, alpha=0.5, label="Non-Setosa", color="green")
plt.axvline(x=2, color='red', linestyle='--', label="Petal Length = 2 cm")

plt.title(f"Distribution of Petal Length with Conditional Probability: P(Setosa | Petal length < 2 cm) = {P_A_given_B:.2f}")
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
